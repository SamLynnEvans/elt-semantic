```python
import duckdb
import pandas as pd

pd.set_option("display.max_colwidth", None)

con = duckdb.connect("/app/database/mydb.duckdb", read_only=True)
```

```python
import duckdb
from IPython.core.magic import register_cell_magic
from IPython.display import display

# Global variable for last DataFrame
LAST_DF = None

@register_cell_magic
def dsql(line, cell):
    global LAST_DF
    query = cell.strip()
    try:
        result = con.sql(query)
        LAST_DF = result.to_df()
        display(LAST_DF)
        return LAST_DF
    except Exception as e:
        print("SQL Error:", e)
```
